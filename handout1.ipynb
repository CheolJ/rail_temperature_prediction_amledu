## **2.5. Define Function ; maximum, minimum and temperature difference**
> 앞으로 코딩하면서 정말 정말 많이 만들게 될 함수부분을 만들어보자!<br>
> 함수는 정말 잘 만들어 놓으면 평-생 써먹을 수 있기 때문에 이번에 한번 빡세게 만들어 보는게 좋을 것 같아서 준비했습니다!! ⊂((・▽・))⊃ <br>
>    - 함수의 목적 : 레일 데이터 자료 한 **행**을 받아 **최대값**과 **최소값**, 그리고 **최대값과 최소값의 차이**를 구하는 함수를 만들어 보자 <br>
>    - 함수의 조건 : 입력 받는 데이터는 레일 데이터 한행만 받음.
>    - **함수의 리턴 값 : 데이터의 최대값, 최소값, 최대값 최소값의 차이**<br>
>    - 함수를 생성한 후, 아래의 문제를 해결하자!


> 문제 : 10분 간격으로 계측된 레일 온도 데이터에 대하여 최대값과 최소값 그리고 그 차이를 구해보자.<br>
>    - 문제 조건 1 : 함수를 생성해서 사용할 것.
>    - 문제 조건 2 : 10분 간격으로 최대값 최소값 그리고 그 차이를 구할 것.

```python
def temp(data_row):
    data_row_max = data_row.max()
    data_row_min = data_row.min()
    data_row_diff = data_row_max - data_row_min
    return [data_row_max, data_row_min, data_row_diff]
```

```python
temp_diff_n = pd.DataFrame(columns=["date","max","min","diff"])
temp_diff_n.date = temp_data_n.date
imsi = temp_data_n.drop(["date"], axis=1)
temp_diff_n[['max', 'min', 'diff']] = imsi.apply(lambda x:pd.Series(temp(x)), axis=1 )
temp_diff_n.head()
```

```python
temp_diff_w = pd.DataFrame(columns=["date","max","min","diff"])
temp_diff_w.date = temp_data_w.date
imsi = temp_data_w.drop(["date"], axis=1)
temp_diff_w[['max', 'min', 'diff']] = imsi.apply(lambda x:pd.Series(temp(x)), axis=1 )
temp_diff_w.head()
```
